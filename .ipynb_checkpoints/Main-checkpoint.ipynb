```python
import pandas as pd
from pymongo import MongoClient
import pymongo
import numpy as np
import matplotlib.pyplot as plt
import math

```

```python
def conn(dbname):
    m_client=pymongo.MongoClient()
    db=m_client[dbname]
    return db
```

```python
def cal_rs(ticker,indx):
    db=conn('StockMarket')
    dfs=db[ticker].find()
    dfs=pd.DataFrame(list(dfs))
    dfi=db[indx].find()
    dfi=pd.DataFrame(list(dfi))
    dfs=dfs[['Date','Close','p_change']]
    dfs['Date']=pd.to_datetime(dfs.Date)
    dfs=dfs.set_index('Date')
    dfi=dfi[['Date','Close','p_change']]
    dfi['Date']=pd.to_datetime(dfi.Date)
    dfi=dfi.set_index('Date')
    return dfs,dfi
    
```

```python
dfs,dfi=cal_rs('ACC','nifty50')
```

```python
dfrs=pd.merge(dfs,dfi,on='Date')
dfrs['rs']=round((dfrs.Close_x/dfrs.Close_y)*100,2)
dfrs['prs']=round((dfrs.p_change_x/dfrs.p_change_y),2)

dfrs.head(20)
```

```python
#plt.figure(figsize=(8,6), dpi=50)
fig, (ax1, ax2,ax3) = plt.subplots(3, sharex=True)
ax1.plot(dfrs.index,dfrs['Close_x'])
ax2.plot(dfrs['rs'])
ax3.plot(dfrs['prs'])
```

```python
def rs_rank(period):
    rank=dfrs
    rank['rnk_max']=rank.rs.rolling(period).max()
    rank['rnk_min']=rank.rs.rolling(period).min()
    rank.dropna(inplace=True)
    rank['rnk']=round((rank.rs-rank['rnk_min'])/(rank['rnk_max']-rank['rnk_min'])*100,2)
    return rank
```

```python
df=rs_rank(250)

```

```python
cor_df=df[['Close_x','rnk']]
```
