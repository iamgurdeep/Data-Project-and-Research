```python
def backtest(df):
    assets=['YesBank']
    for asset in assets:
        globals()[asset]=pd.DataFrame(df.index).assign(ACash=0,Trade_Count=0,PositionDays=0,BuyPrice=0,ABuyPrice=0,SellPrice=0,CurrentPrice=0,Quantity=0,DailyPNL=0,DailyPerChange=0.0)
        globals()[asset].set_index('Date',inplace=True)
        globals()[asset].iloc[0,0]=100000
    pos=0
    days_in_pos=0
    buy_price=0
    sell_price=0
    trade_results=[[]]
    for i in df.index:
        if i != df.index[0]:
            buy= ((df['pos'][i]== 11) and (df['pos'].shift(1)[i]==1)) or ((df['pos'][i]== 11) and (df['pos'].shift(1)[i]==0))
            sell= ((df['pos'][i]== 0) and (df['pos'].shift(1)[i]==1)) or ((df['pos'][i]== 0) and (df['pos'].shift(1)[i]==11))


            if buy: # Buy Triggred
                if pos==0:
                    pos=1
                    current_price=df['Close'][i]
                    buy_price=current_price
                    YesBank.at[i,'Trade_Count']=1
                    print(f'Buy Triggred,Buying at {current_price} on {i}')

                else : # Already in buy Position Increase the number of days in trade and note today's % Change
                    days_in_pos+=1
                    current_price=df['Close'][i]
                    YesBank.at[i,'Trade_Count']=1
                    YesBank.at[i,'PositionDays']=days_in_pos

                    Today_Change=df['P_Daily_Change'][i]
                    YesBank.at[i,'DailyPerChange']=Today_Change
                    print(f' Buy Triggred but Aready in buy trade on {i}')

            elif sell: # Sell Triggred Close the Position 
                if pos==1:                    
                    pos=0
                    current_price=df['Close'][i]
                    sell_price=current_price
                    YesBank.at[i,'Trade_Count']=0
                    days_in_pos=0
                    print(f'Sell Triggred,Selling at {current_price} on {i}')
                    gl=((sell_price/buy_price)-1)
                    trade_results.append(gl)
                else: ## Sell Triggred but not in buy postion at this moment so do nothing 
                    print(f' Sell Triggred but no trade is ON  on {i}')
                    pass

            else :#or sell: ##No Buy or Sell Triggred but trade could be already in Place
                if pos==1: # Buy Trade is active ,Increase the number of days in trade and note today's % Change
                    days_in_pos+=1
                    current_price=df['Close'][i]
                    YesBank.at[i,'Trade_Count']=1
                    YesBank.at[i,'PositionDays']=days_in_pos

                    Today_Change=df['P_Daily_Change'][i]
                    YesBank.at[i,'DailyPerChange']=Today_Change
                    print(f' Continue -Aready in buy trade on {i}')
                else: # Not in trade to Do nothing 
                    print(f' Nothing Triggred and no trade is ON  on {i}')
                    print(f"Current Close{df['Close'][i]} and Last High to Corss {df['High'].shift(1)[i]}")
                    pass
    return YesBank,trade_results
```

```python
def matrix(trade_result):
    matrixList=[]
    gains=0
    ng=0
    losses=0
    nl=0
    for i in trade_result:
        if(i>0):
            gains+=i
            ng+=1
        else:
            losses+=i
            nl+=1

    if(ng>0):
        avgGain=gains/ng
        maxR=max(trade_result)
    else:
        avgGain=0
        maxR=0

    if(nl>0):
        avgLoss=losses/nl
        maxL=min(trade_result)
        ratio=-avgGain/avgLoss
    else:
        avgLoss=0
        maxL=0
        ratio=999

    if(ng>0 or nl>0):
        battingAvg=(ng/(ng+nl))*100
    else:
        battingAvg=0
    #### Calculate Concurrent Loss and Gain 
    gainlist=[]
    losslist=[]
    npg=0
    npl=0
    for i in trade_result:
        if i!=trade_result[-1]:
            if i>0:
                losslist.append(npl)
                npg+=1
                npl=0
            elif i<0:
                gainlist.append(npg)
                npl+=1
                npg=0
        else:
            if i>0:
                npg+=1
            if i<0:
                npl+=1
            losslist.append(npl)
            gainlist.append(npg)
    
    total=ng+nl
    winavg=round(battingAvg,2)
    profit_loss_ratio=round(ratio,2)
    maxprofit=round(max(trade_result)*100,2)
    maxloss=round(min(trade_result)*100,2)
    avggain_trade=round(avgGain*100,2)
    avgloss_trade=round(avgLoss*100,2)
    max_gain_streak=max(gainlist)
    max_loss_streak=max(losslist)
    matrixList.extend([total,ng,nl,winavg,profit_loss_ratio,maxprofit,maxloss,avggain_trade,avgloss_trade,max_gain_streak,max_loss_streak])
    
    return matrixList
    
    # print(f'Total Trades : {ng+nl}')
    # print(f'Number of Winners: {ng} and Number of Loss Trades : {nl}')
    # print(f'Win Average: {round(battingAvg,2)}%')
    # print(f'Profil/Loss Ratio : {round(ratio,2)}')
    # print(f'maximum % Profit in one trade : {round(max(trade_result)*100,2)}%')
    # print(f'maximum % Loss in one trade : {round(min(trade_result)*100,2)}%')
    # print(f'Averge gain per Trade: {round(avgGain,2)}')
    # print(f'Averge loss per Trade: {round(avgLoss,2)}')
```

```python
#Download Data from yahoo
### TEST Scnarios - TCS- Bull ,ACC- modrate up (Sidewase),YESBank- Back to Zero ,ITC(stock dosnt even Move)
def getdata(asset,start,end):
    stock=asset
    indexes=['^NSEI']
    if stock not in indexes:
        startdate=start
        # enddate='2020-01-01'
        enddate=end
        exchange=['.NS','.BO']
        dfo = yf.download(stock+exchange[0],start=startdate,end=enddate)
    else:

        startdate=start
        # enddate='2020-01-01'
        enddate=end
        dfo = yf.download(stock,start=startdate,end=enddate)
    return dfo
```

```python
##[pos]-->11 buy
##[pos]-->1 Continue
##[pos]-->0 Sell
def conditions(df,buycheck,slcheck,conditioncheck):
    
    ########Buy In strength##########
    # df['In']=np.where(df['Close'].shift(1)>df['High'].shift(2).rolling(84).max(),1,0)
    # df['Out']=np.where(df['Close']<df['Low'].shift(1).rolling(21).min(),0,1)
    # df['In']=np.where((df['Close']>df['High'].shift(1).rolling(21).max()),1,0)
    # df['Out']=np.where(df['Close']<df['Low'].shift(1).rolling(10).min(),0,1)
    ###############Best for Yesbank##################always keep Out at lower then BuY Days####21,21
    # Stock that dont move much like ITC need Less Days like 5 .. 
    df['In']=np.where((df['Close']>df['High'].shift(buycheck)) & (df['High'].shift(buycheck)<df['High'].shift(3)) , 1,0)
    ## Set extra condition  & (df['Close'].shift(1)>df['Close'].rolling(conditioncheck).mean())
    df['Out']=np.where((df['Close']<df['Low'].shift(slcheck)),0,1)



    for i in df.index:
        df.at[i,'pos']=str(df['In'][i])+str(df['Out'][i])
    df['pos']=df['pos'].astype(str).astype(int)
    return df
```

```python
startDate=['2007-01-01','2008-01-01']
EndDate=['2008-01-01','2009-01-01']
assets=['TCS','WIPRO','HDFCBANK','ICICIBANK','RELIANCE','BAJFINANCE','JSWSTEEL']
matrixdf=pd.DataFrame(index=['period','total_trades','gain_count','loss_count','winavg','profit_loss_ratio','maxprofit',
                                                    'maxloss','avggain_trade','avgloss_trade','max_win_streek','max_loss_streek','StratergyReturns','UnderLineReturns'])
for j in range(len(startDate)):
    
    for asset in assets:
        dfo=getdata(asset=asset,start=startDate[j],end=EndDate[j])
        buycheck=[1]
        slcheck=[1]
        conditioncheck=[50]
        matrixList=[]
        for i in buycheck:
            df=dfo.copy()
            df['P_Daily_Change']=df['Close'].pct_change().apply(lambda x: round(x,2))
            df.dropna(inplace=True)
            ## Calling conditions
            df=conditions(df=df,buycheck=i,slcheck=slcheck[0],conditioncheck=conditioncheck[0])

            # Calling backtester
            YesBank,trade_results=backtest(df)
            YesBank['CReturns']=(((1+YesBank['DailyPerChange']).cumprod()-1)*100).apply(lambda x:round(x,2))
            df['CReturns']=(((1+df['P_Daily_Change']).cumprod()-1)*100).apply(lambda x:round(x,2))
            df['S_DailyChange']=YesBank['DailyPerChange']
            df['S_CReturns']=YesBank['CReturns']

            S_CReturns=YesBank[-1:]['CReturns'][0]
            U_CReturns=df[-1:]['CReturns'][0]
            
            mylist=matrix(trade_results)
            matrixList.append(f'{startDate[j]}-{EndDate[j]}')
            
            matrixList.extend(mylist)
            matrixList.extend([S_CReturns,U_CReturns])
            matrixdf[asset]=matrixList
```
