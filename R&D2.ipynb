```python
import pandas as pd
import json
import numpy as np
import datetime
from pymongo import MongoClient
import Orange
import matplotlib.pyplot as plt

# Create Mongo client instance
mongo_client=MongoClient()
```

```python
optdb=mongo_client['NSEOPTIDX']
#Fetching Data from DB

optdata=optdb['NIFTY'].find()
optdf=pd.DataFrame(list(optdata))
optdf.drop('_id',axis=1,inplace=True)
```

```python
def get_opt_data():
    dfc=optdf.copy()
    dfc=dfc[['Datetime','EXPIRY_DT','OPTION_TYP','STRIKE_PR','SETTLE_PR','OPEN_INT', 'CHG_IN_OI']]
    dfc=dfc.rename(columns={'SETTLE_PR':'CLOSE'})
    call_df=dfc.loc[dfc['OPTION_TYP']=='CE']
    call_df['DAYS_LEFT']=((call_df['EXPIRY_DT']-call_df['Datetime']).dt.days)+1
    call_df['CALL_VALUE']=call_df['CLOSE']/call_df['DAYS_LEFT']
    put_df=dfc.loc[dfc['OPTION_TYP']=='PE']
    put_df['DAYS_LEFT']=((put_df['EXPIRY_DT']-put_df['Datetime']).dt.days)+1
    put_df['PUT_VALUE']=put_df['CLOSE']/put_df['DAYS_LEFT']
    return call_df,put_df
```

```python
def checkit(exp_date,strike):
    call_df,put_df=get_opt_data()
    call_df.set_index('Datetime',inplace=True)
    call_df=call_df.loc[(call_df['EXPIRY_DT']==exp_date)&(call_df['STRIKE_PR']==strike)]
    put_df.set_index('Datetime',inplace=True)
    put_df=put_df.loc[(put_df['EXPIRY_DT']==exp_date)&(put_df['STRIKE_PR']==strike)]
    return call_df,put_df
    
```

```python
call_df,put_df=checkit('2020-06-25',9600)

```
